```python
import pandas as pd
```

```python
athletes = pd.read_csv("athlete_events_post_1990.csv")
athletes.info()
athletes = athletes.rename(columns={"City":"GamesLocation","Team":"CountryRepresented"})
athletes.head()
#Sex, Medal
#Object --> String
```

```python
athletes["Sex"] = athletes["Sex"].astype("category")
athletes["Medal"] = athletes["Medal"].astype("category")
columns = ["Name","CountryRepresented","NOC","Games","Season","GamesLocation","Sport","Event"]
athletes[columns] = athletes[columns].astype("string")
athletes.head()
```
